```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
print(iris['target_names'])
```

```python
## independent features
x = pd.DataFrame(iris['data'],columns = ['sepal length','sepal width','petal length','petal width'])
```

```python
## dependent feature
y = iris['target']
```

```python
## train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
## apply decisiontree clasifier 
from sklearn.tree import DecisionTreeClassifier
tree_classifier =  DecisionTreeClassifier()
# tree_classifier =  DecisionTreeClassifier(max_depth=2)
```

```python
tree_classifier
```

```python
tree_classifier.fit(x_train,y_train)
```

```python
## visualize the Decision tree
from sklearn import tree
plt.figure(figsize=(18,22))
tree.plot_tree(tree_classifier,filled=True)
```

```python
y_pred =  tree_classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm = confusion_matrix(y_test,y_pred)
report = classification_report(y_test,y_pred)
```

```python
print("confusion matrix is ")
print(cm)
print("Classification report is ")
print(report)
```

## Decision Tree Classifier Prepruning And Hyperparameter Tuning



```python
param = {
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import  GridSearchCV
```

```python
tree_model = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(tree_model,param_grid=param,cv=5,scoring='accuracy')
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred_cv = grid.predict(x_test)
```

```python
matrix_cv = confusion_matrix(y_test,y_pred_cv)
report_cv = classification_report(y_test,y_pred_cv)
```

```python
print("confusion matrix is ")
print(matrix_cv)
print("Classification report is ")
print(report_cv)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score(y_test,y_pred_cv)
```

```python
print("accuracy score for grid is ",score)
```
