## Local Outlier Implemenatation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_circles
from sklearn.neighbors import LocalOutlierFactor
```

```python
x,y = make_circles(n_samples=750,factor=0.3,noise=0.2)
```

```python
df = pd.DataFrame(x)
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c=y)
```

```python
model_LocaloutlierFactor = LocalOutlierFactor(n_neighbors=5,contamination=0.2)
```

```python
y_pred = model_LocaloutlierFactor.fit_predict(x)
```

```python
scores = model_LocaloutlierFactor.negative_outlier_factor_
```

## plotting datset  with pointing out outliers

```python
plt.title("Local Outlier Factor (LOF)")
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], color='k', s=3., label='Data points')
# plot circles with radius proportional to the outlier scores
radius = (scores.max() - scores) / (scores.max() - scores.min())
plt.scatter(df.iloc[:, 0].values, df.iloc[:, 1].values, s=50 * radius, edgecolors='r',
            facecolors='none', label='Outlier scores')
plt.show()
```
