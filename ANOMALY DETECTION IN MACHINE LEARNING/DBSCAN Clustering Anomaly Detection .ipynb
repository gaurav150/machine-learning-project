```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
```

```python
x,y = make_circles(n_samples=750,factor=0.3,noise=0.1)
```

```python
x
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
model_dbscan = DBSCAN(eps=0.1)
```

```python
model_dbscan.fit_predict(x)
```

```python
model_dbscan.labels_
```

## plotting datset  with pointing out outliers

```python
plt.scatter(x[:,0],x[:,1],c=model_dbscan.labels_)
```

## plotting Original dataset plot using scatter plot

```python
plt.scatter(x[:,0],x[:,1],c=y)
```
