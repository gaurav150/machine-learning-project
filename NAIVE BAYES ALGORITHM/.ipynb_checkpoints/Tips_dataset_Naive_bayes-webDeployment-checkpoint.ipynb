```python
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
## Onehot encoding--- ColumnTrnasformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import sys
import numpy as np
import pickle as pk
```

```python
## independent and dependent features 
x = df[['total_bill','tip', 'sex', 'smoker', 'day', 'size']]
y = df['time']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
x_train.head()
```

## Feature Encoding(Label Encoding And Onehot Encoding)

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
x_train['sex']=le1.fit_transform(x_train['sex'])
x_train['smoker']=le2.fit_transform(x_train['smoker'])
y_train = le3.fit_transform(y_train)
```

```python
x_test['sex']=le1.transform(x_test['sex'])
x_test['smoker']=le2.transform(x_test['smoker'])
y_test = le3.transform(y_test)
```

```python
x_test.head()
```

```python
ct=ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[4])],
                                   remainder='passthrough')
```

```python
np.set_printoptions(threshold=sys.maxsize)
x_train=ct.fit_transform(x_train)
```

```python
x_test = ct.transform(x_test)
```

```python
x_test
```

```python
gnb_tips = GaussianNB()
gnb_tips
```

```python
gnb_tips.fit(x_train,y_train)
```

```python
y_pred = gnb_tips.predict(x_test)
```

```python
score_tips = accuracy_score(y_test,y_pred)
report_tips = classification_report(y_test,y_pred)
matrix_tips = confusion_matrix(y_test,y_pred)
```

```python
print("confusion matrix is ")
print(matrix_tips)
print("classification report is ")
print(report_tips)
print("accuracy score is ",score_tips)
```

```python
pk.dump(le1,open('le1.pkl','wb')) ## labelEncoder model 
pk.dump(le2,open('le2.pkl','wb')) ## labelEncoder model 
pk.dump(le3,open('le3.pkl','wb')) ## labelEncoder model 

pk.dump(ct,open('Column_transformer.pkl','wb'))  ## columnTransformer model 

pk.dump(gnb_tips,open('gaussian_tips.pkl','wb'))   ## gaussian_tips model
```
