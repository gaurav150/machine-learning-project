## Naive Baye's Machine Learning Algorithms Implementation

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
x,y = load_iris(return_X_y=True)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb
```

```python
gnb.fit(x_train,y_train)
```

```python
y_pred = gnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
score = accuracy_score(y_test,y_pred)
report = classification_report(y_test,y_pred)
matrix = confusion_matrix(y_test,y_pred)
```

```python
print("confusion matrix is ")
print(matrix)
print("classification report is ")
print(report)
print("accuracy score is ",score)
```

```python
## independent and dependent features 
x1 = df[['total_bill','tip', 'sex', 'smoker', 'day', 'size']]
y1 = df['time']
```

```python
## train test split
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y1,test_size=0.25,random_state=10)
```

```python
x1_train.head()
```

## Feature Encoding(Label Encoding And Onehot Encoding)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
import warnings
warnings.filterwarnings('ignore')
x1_train['sex']=le1.fit_transform(x1_train['sex'])
x1_train['smoker']=le2.fit_transform(x1_train['smoker'])
y1_train = le3.fit_transform(y1_train)
```

```python
x1_test['sex']=le1.transform(x1_test['sex'])
x1_test['smoker']=le2.transform(x1_test['smoker'])
y1_test = le3.fit_transform(y1_test)
```

```python
x1_test.head()
```

```python
## Onehot encoding--- ColumnTrnasformer

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[4])],
                                   remainder='passthrough')
```

```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
x1_train=ct.fit_transform(x1_train)
```

```python
x1_test = ct.transform(x1_test)
```

```python
x1_test
```

```python
gnb_tips = GaussianNB()
gnb_tips
```

```python
gnb_tips.fit(x1_train,y1_train)
```

```python
y_pred_tips = gnb_tips.predict(x1_test)
```

```python
score_tips = accuracy_score(y1_test,y_pred_tips)
report_tips = classification_report(y1_test,y_pred_tips)
matrix_tips = confusion_matrix(y1_test,y_pred_tips)
```

```python
print("confusion matrix is ")
print(matrix_tips)
print("classification report is ")
print(report_tips)
print("accuracy score is ",score_tips)
```
